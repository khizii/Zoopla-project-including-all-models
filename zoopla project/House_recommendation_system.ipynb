```python
def  remove_outlier_sqft(df):
    df_output=pd.DataFrame()
    for key,subdf in data.groupby('location'):
        m=np.mean(subdf.price_per_sqfeet)
        st=np.std(subdf.price_per_sqfeet)
        gen_df=subdf[(subdf.price_per_sqfeet>(m-st))&(subdf.price_per_sqfeet <= (m+st))]
        df_output=pd.concat( [df_output , gen_df ], ignore_index=True)
    return df_output

data=remove_outlier_sqft(data)
data.describe()
```

```python
import nltk
from nltk.stem import PorterStemmer
nltk.download('punkt')
stemmer = PorterStemmer()
```

```python
data['property_type'] = data['property_type'].str.lower()
data['location'] = data['location'].str.lower()
data['city'] = data['city'].str.lower()
```

```python
data['location'] = data['location'].apply(lambda x: ' '.join([stemmer.stem(word) for word in nltk.word_tokenize(x)]))
data['city'] = data['city'].apply(lambda x: ' '.join([stemmer.stem(word) for word in nltk.word_tokenize(x)]))
```

```python
data.head()
```

```python
data['location'] = data['location'] + ' , ' + data['city']
data.drop(['price_per_sqfeet'], axis=1, inplace=True)
data.head()
```

```python
data = data.sample(frac=1, random_state=42)
```

```python
# Separate data for sale and for rent
sale_data = data[data['purpose'] == 'For Sale']
rent_data = data[data['purpose'] == 'For Rent']
```

```python
# Delete the last  rows from the DataFrame
sale_data = sale_data.iloc[:-60000]
rent_data=rent_data.iloc[:-20000]
sale_data = sale_data.reset_index(drop=True)
rent_data = rent_data.reset_index(drop=True)
```

```python
sale_data.shape
rent_data.shape
```

```python
# text vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
sale_data.head()
```

```python
sale_vectors=cv.fit_transform(sale_data['location']).toarray()
rent_vectors=cv.fit_transform(rent_data['location']).toarray()
```

```python
sale_vectors
```

```python
input1=input("Enter your choice : (for sale or for rent): ")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sale_similarity=cosine_similarity(sale_vectors)
rent_similarity=cosine_similarity(rent_vectors)
```

```python
sale_similarity.shape
```

```python
rent_similarity.shape
```

```python
sale_data.tail()
```

```python
def recommend(location):
  if input1=='for sale':
    cities=['karachi','lahor','islamabad','faisalabad','rawalpindi']
    if location in cities:
      movie_index = sale_data[sale_data['city']== location].index[0]
      distances = sale_similarity[movie_index]
      movies_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])[:30]
    else:
      movie_index = sale_data[sale_data['location']== location].index[0]
      distances = sale_similarity[movie_index]
      movies_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])[:30]
    print("Recommended property are : ")
    for i in movies_list:
        recommended_property=[]
        recommended_property_1 = sale_data.iloc[i[0]]['property_type']
        recommended_property_2 = sale_data.iloc[i[0]]['location']
        recommended_property_3 = sale_data.iloc[i[0]]['baths']
        recommended_property_4 = sale_data.iloc[i[0]]['bedrooms']
        recommended_property_5 = sale_data.iloc[i[0]]['Total_Area']
        recommended_property_6 = sale_data.iloc[i[0]]['price']
        recommended_property.append({
          "property_type": recommended_property_1,
          "location": recommended_property_2,
          "baths": recommended_property_3,
          "bedrooms": recommended_property_4,
          "Total_Area":recommended_property_5,
          "price":recommended_property_6
      })
        print(recommended_property)
  elif input1=='for rent':
    cities=['karachi','lahor','islamabad','faisalabad','rawalpindi']
    if location in cities:
      movie_index = rent_data[rent_data['city']== location].index[0]
      distances = rent_similarity[movie_index]
      movies_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])[:30]
    else:
      movie_index = rent_data[rent_data['location'] == location].index[0]
      distances = rent_similarity[movie_index]
      movies_list = sorted(enumerate(distances), reverse=True, key=lambda x: x[1])[:30]
    print("Recommended property are: ")
    for i in movies_list:
          recommended_property=[]
          recommended_property_1 = rent_data.iloc[i[0]]['property_type']
          recommended_property_2 = rent_data.iloc[i[0]]['location']
          recommended_property_3 = rent_data.iloc[i[0]]['baths']
          recommended_property_4 = rent_data.iloc[i[0]]['bedrooms']
          recommended_property_5 = rent_data.iloc[i[0]]['Total_Area']
          recommended_property_6 = rent_data.iloc[i[0]]['price']
          recommended_property.append({
          "property_type": recommended_property_1,
          "location": recommended_property_2,
          "baths": recommended_property_3,
          "bedrooms": recommended_property_4,
          "Total_Area":recommended_property_5,
          "price":recommended_property_6
      })
          print(recommended_property)
```
